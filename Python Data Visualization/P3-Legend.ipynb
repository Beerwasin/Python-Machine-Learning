```python
import matplotlib.pyplot as plt

day=[1,2,3,4,5,6,7]
temp2561=[20,35,28,30,18,20,18]
temp2560=[15,25,38,30,26,27,12]
temp2559=[20,17,19,20,25,30,40]

plt.plot(day,temp2561,color='red',linewidth=5,label='2561')  #label ตัวสุดท้ายคือตั้งชื่อให้ legend
plt.plot(day,temp2560,color='blue',linewidth=5,label='2560') 
plt.plot(day,temp2559,color='black',linewidth=5,label='2559') 
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.legend()
plt.grid()  #ใส่ตารางพื้นหลัง
plt.show()
```
