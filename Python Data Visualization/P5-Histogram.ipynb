```python
import matplotlib.pyplot as plt

bloodx=[113,90,85,150,105,145]
bloody=[110,75,80,120,60,125]
plt.hist([bloodx,bloody],bins=2,color=['red','blue'],label=['men','women'])
#plt.hist([bloodx,bloody],bins=2,color=['red','blue'],label=['men','women'],orientation='horizontal')  #เปลี่ยนเป็นแนวนอน
plt.xlabel('Blood Sugar Range')    #ระดับน้ำตาลในเลือด
plt.ylabel('Patients')     #จำนวนผู้ป่วย
plt.title('Blood Sugar')   #ชื่อหัวข้อ
plt.legend()

#ใช้แบ่งช่วงความถี่ของข้อมูลแบ่งออกเป็นช่วง,กลุ่ม
```
