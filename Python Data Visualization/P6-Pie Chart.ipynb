```python
import matplotlib.pyplot as plt

text=['internet','phone','food','car','coffee']
money=[500,200,2500,1500,700]
plt.pie(money,labels=text,radius=1.5,autopct='%0.0F%%',shadow=True,explode=[0.2,0,0,0,0],startangle=180)               # radius=ใช้ปรับขนาด autopct=ใช้แสดงค่า %  explode=ใช้แยกชิ้นส่วน 
#plt.axis('equal')              #ใช้ปรับขนาดให้สมส่วน
plt.show()
```
